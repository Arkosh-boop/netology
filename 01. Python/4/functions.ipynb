Вы работаете секретарем и вам постоянно приходят различные документы. Вы должны быть очень внимательны, чтобы не потерять ни один документ. Каталог документов хранится в следующем виде:

Перечень полок, на которых находятся документы хранится в следующем виде:

Необходимо реализовать пользовательские команды (внимание! это не название функций, которые должны быть выразительными, а команды, которые вводит пользователь, чтобы получить необходимый результат):
  
- p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;  
- s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится  
*Правильно обработайте ситуации, когда пользователь будет вводить несуществующий документ*.
- l – list – команда, которая выведет список всех документов в формате   
**passport "2207 876234" "Василий Гупкин"**  
- as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень. *Предусмотрите случай, когда пользователь добавляет полку, которая уже существует*.
- d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок. *Предусмотрите сценарий, когда пользователь вводит несуществующий документ*;  
- m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую. *Корректно обработайте кейсы, когда пользователь пытается переместить несуществующий документ или переместить документ на несуществующую полку*;  
- a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться. *Корректно обработайте ситуацию, когда пользователь будет пытаться добавить документ на несуществующую полку*.

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'},
    {'type': 'insurance', 'number': '106', 'name': 'Абу Иванов'}
]


directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': ['106']
}


# p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;
def p_people():
    number_doc = input("Введите номер документа:\n")
    i = 0
    for document in documents:
        i += 1
        if document['number'] == number_doc:
            print(f'Владелец документа: {document["name"]} \n')
            break
        elif i == len(documents):
            print('Ошибка! Такого номера документа нет в базе. Повторите операцию.\n')


# s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится
def s_shelf():
    while True:
        number_doc = input('Введите номер документа:\n')
        i = 0
        for directory in directories.items():
            i += 1
            if number_doc in directory[1]:
                print(f"Документ {number_doc} находится на полке {directory[0]}\n")
                return
            elif i == len(directories):
                print(f'Ошибка! Документ {number_doc} на полках не найден. Повторите операцию. \n')
                continue


# l – list – команда, которая выведет список всех документов в формате
# passport "2207 876234" "Василий Гупкин"
def l_list():
    for documents_list in documents:
        print(f"{documents_list['type'], documents_list['number'], documents_list['name']}")


# d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок.
def d_delete():
    number_doc = input('Введите номер документа который хотите удалить:\n')
    for document in documents:
        if document['number'] == number_doc:
            document.clear()
            for directory in directories.items():
                if number_doc in directory[1]:
                    directory[1].remove(number_doc)
                    print(f'Документ "{number_doc}" успешно удалён.')
            break
    else:
        print('Ошибка! Документ с таким номером не найден. Придется вызвать функцию еще раз.\n')


# m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую.
def m_move():
    doc_number = input('Введите номер документа:\n')
    move_to = input('Введите номер полки куда необходимо добавить документ:\n')
    for shelf, item in directories.items():
        if doc_number in item and move_to in directories.keys():
            directories[shelf].remove(doc_number)
            directories[move_to].append(doc_number)
            print('Документ "{}" перемещен на полку {}'.format(doc_number, move_to))
            return
    print('\nНеверные вводные данные документа или полки, пожалуйста, проверьте и введите заново!\n')


def a_add():
    input_type = input('Введите тип документа:\n')
    input_number = input('Введите номер документа:\n')
    input_name = input('Введите имя владельца документа:\n')
    input_shelf = input('Введите номер полки:\n')
    new_dict = dict(type=input_type, number=input_number, name=input_name)
    documents.append(new_dict)
    for _ in directories.items():
        if input_shelf.isdigit() and input_shelf in directories:
            directories[input_shelf].append(input_number)
            print('Документ добавлен в каталог:\n {} {} {}'.format(new_dict['type'], new_dict['number'],
                                                                   new_dict['name']))
            print('Полка:', input_shelf)
            break

        else:
            print('Ошибка! Полки с таким номером нет. \nПридется вызвать функцию еще раз.\n')
            break


# as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень. Полки могут идти не подряд
def add_shelf():
    while True:
        shelf_number = input("Введите номер новой полки:\n")
        if shelf_number.isdigit():
            for item in directories.items():
                if int(shelf_number) != item[0]:
                    directories.update({shelf_number: []})
                    print('Полка:', shelf_number, 'успешно добавлена в перечень.\n')
                    return
                elif int(shelf_number) <= (len(directories)):
                    print('Ошибка! Полка с таким номером существует. Повторите операцию.\n')
        else:
            print('Ошибка! Номер полки может быть обозначен только цифрами. Повторите ввод\n')


def h_help():
    print('Добро пожаловать в гости к секретарю.\nВыберите одну из команд: \n'
          'p - Узнать кому принадлежит документ; \n'
          'l - Вывести список документов в формате "passport "2207 876234" "Василий Гупкин"; \n'
          's - Узнать номер полки, на которой хранится документ; \n'
          'a - Добавить новый документ; \n'
          'd - Удалить документ из каталога и из перечня полок; \n'
          'm - Перемещение документа на другую полку; \n'
          'as - Команда, которая спросит номер новой полки и добавит ее в перечень; \n'
          'x - Выход;\n'
          '? - Вызов этой справки\n')


h_help()


while True:
    user_start = input('Пожалуста введите команду из списка, соблюдая регистр. \n x - закрыть, ? - справка \n')
    if user_start == 'x':
        print('на этом все')
        break
    elif user_start == 'p':
        p_people()
    elif user_start == 's':
        s_shelf()
    elif user_start == 'l':
        l_list()
    elif user_start == 'd':
        d_delete()
    elif user_start == 'm':
        m_move()
    elif user_start == 'a':
        a_add()
    elif user_start == 'as':
        add_shelf()
    elif user_start == '?':
        h_help()
    else:
        print('Ваша команда не доступна, проверьте доступные команды используя "?"')
```
